```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('heart_2020_cleaned.csv')
df
```

```python
na_values = df.isna().sum()
print(na_values)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
df['GenHealth'] = ordinal_encoder.fit_transform(df[['GenHealth']])
df 
```

```python
df['Smoking'] = ordinal_encoder.fit_transform(df[['Smoking']])
df['AlcoholDrinking'] = ordinal_encoder.fit_transform(df[['AlcoholDrinking']])
df['Stroke'] = ordinal_encoder.fit_transform(df[['Stroke']])
df['DiffWalking'] = ordinal_encoder.fit_transform(df[['DiffWalking']])
df['Sex'] = ordinal_encoder.fit_transform(df[['Sex']])
df['Race'] = ordinal_encoder.fit_transform(df[['Race']])
df['Diabetic'] = ordinal_encoder.fit_transform(df[['Diabetic']])
df['PhysicalActivity'] = ordinal_encoder.fit_transform(df[['PhysicalActivity']])
df['Asthma'] = ordinal_encoder.fit_transform(df[['Asthma']])
df['KidneyDisease'] = ordinal_encoder.fit_transform(df[['KidneyDisease']])
df['SkinCancer'] = ordinal_encoder.fit_transform(df[['SkinCancer']])

df['HeartDisease'] = ordinal_encoder.fit_transform(df[['HeartDisease']])
df
```

```python
df['AgeCategory'] = df['AgeCategory'].str.replace('80 or older', '80-100')
df['AgeCategory'] = df['AgeCategory'].apply(lambda x: (int(x.split('-')[0]) + int(x.split('-')[1])) / 2)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

X = df.iloc[:, 1:]
y = df.iloc[:, 0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy
```
