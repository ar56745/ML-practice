```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('HousingPrices-Amsterdam-August-2021.csv', index_col=0)
df
```

```python
new_df = df.drop('Price',axis='columns')
new_df = new_df.drop('Address',axis='columns')
new_df = new_df.drop('Lon',axis='columns')
new_df = new_df.drop('Lat',axis='columns')
new_df = new_df.drop('Zip',axis='columns')
new_df.drop(df.tail(4).index, inplace = True)
new_df=new_df.dropna()
print(new_df.isnull().sum())
new_df
```

```python
price_df = df['Price']
price_df=price_df.dropna()
print(price_df.isnull().sum())
price_df
```

```python
reg=linear_model.LinearRegression()
reg.fit(new_df,price_df)
```

```python
reg.predict([[200, 4]])
```
